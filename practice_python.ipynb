```python
students = [
    {"name": "Alice", "scores": [85, 90, 78]},
    {"name": "Bob", "scores": [70, 88, 82]},
    {"name": "Charlie", "scores": [95, 92, 91]}
]
print(students)
for student in students:
    print(student['name'], student['scores'])
```

Loooping through list of dictionaries

```python
for person in students:
    print(person['name'], end=' ')

print('\n')

for person in students:
    print(person['name'], person end='\n')
```

```python
#print out each student average score

for student in students:
    name = student['name']
    scores = student['scores']

    avg = sum(scores) / len(scores)

    print(f'{name} has average score: {avg:.2f}') 
```

Getting the letter grade    

```python
for student in students:
    name = student['name']
    scores = student['scores']

    avg = sum(scores) / len(scores)
    
    if avg >= 90:
        grade = 'A'
    elif avg >= 80:
        grade = 'B'
    elif avg >= 70:
        grade = 'C'
    elif avg >= 60:
        grade = 'D'
    else:
        grade = 'F'
    
    print(f'{name}: {avg:.2f} ➡️ Grade {grade}', '\n')
```
